```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv(r'C:\Users\Abhishek\Documents\movies.csv')
```

```python
movies_data.head()
#top 5 movies printing
```

```python
#chcking row and column in movies data
movies_data.shape
```

```python
#content similarity similaruty chcking for some movies

selected_similarity=['genres','keywords','tagline','director','cast']
```

```python
#if there any null value in selected_similarity they willb replaced by string
for similarity in selected_similarity:
    movies_data[similarity]=movies_data[similarity].fillna('')
```

```python
#combining all similaritiee in one string
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['director']+' '+movies_data['cast']
print(combined_features)
```

```python

vectorizer = TfidfVectorizer()
```

```python
# converting the text data to numerival values
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#similar using  cosine_similarity
similar_value=cosine_similarity(feature_vectors)
print(similar_value)
```

```python
movie_name = input(' Enter your favourite movie name')
```

```python
list_of_all_movies=movies_data['title'].tolist()
print(list_of_all_movies)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similar_value[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
